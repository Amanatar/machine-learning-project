```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data = credit_card_data.dropna()
```

```python
credit_card_data["Class"].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby("Class").mean()
```

```python
legit_sample =legit.sample(n=200)
```

```python
new_datset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_datset.head()
```

```python
new_datset.tail()
```

```python
new_datset["Class"].value_counts()
```

```python
new_datset.groupby("Class").mean()
```

```python
x = new_datset.drop(columns="Class", axis=1)
y = new_datset["Class"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy on Training data : ", training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy on testing data : ", testing_data_accuracy)
```
